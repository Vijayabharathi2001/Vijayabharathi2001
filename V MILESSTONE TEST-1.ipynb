```python
#find=iqr              #using quantile for removing and detectioning those outlayers
r=b.select_dtypes(exclude=['object'])
r
```

```python
q1=r.quantile(0.25)
q1
```

```python
q3=r.quantile(0.75)
q3
```

```python
IQR=q3-q1
IQR
```

```python
a=((r<q1-1.5*IQR)|(r>q1+1.5*IQR))
a
```

```python
filter=b[~((r<q1-1.5*IQR)|(r>q1+1.5*IQR)).any(axis=1)]
filter
```
